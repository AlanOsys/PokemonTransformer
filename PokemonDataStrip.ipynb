```python
Names = pd.read_csv('Datasets/Common_Spreads.csv',delimiter=',', header=None, skiprows=0,names=['Nature','IVs','Use_Percentage','Name'])
Types = pd.read_csv('Datasets/Typings.csv',delimiter=',', header=None, skiprows=0,names=['Name','Type'])
Items = pd.read_csv('Datasets/Items_May_Use.csv',delimiter=',', header=None, skiprows=0,names=['Item','Percentage','Name'])
Spreads = pd.read_csv('Datasets/Common_Spreads.csv',delimiter=',', header=None, skiprows=0,names=['Nature','IVs','Use_Percentage','Name'])
Moves = pd.read_csv('Datasets/Move_Percentages.csv',delimiter=',', header=None, skiprows=0,names=['Move','Use_Percentage','Pokemon','Name'])
Abilities = pd.read_csv('Datasets/Abilities_Percentages.csv',delimiter=',', header=None, skiprows=0,names=['Ability','Use_Percentage','Pokemon'])
AllMoves = pd.read_csv('Datasets/df_moves.csv',delimiter=',', header=None, skiprows=0,names=['Name','Introducted_in','Type','Power','Acc.','PP','Damage_class','Effect','Prob. (%)'])
f = open("PokemonStatsFile.txt","w",encoding="utf-8")
NameArr = []
for index, NameRow in Names.iterrows():
    currentName = str(NameRow['Name'])
    #print(currentName)
    if(currentName not in NameArr):
        #NameArr.append(currentName)
        PokemonString = ""
        currentName = str(NameRow['Name'])
        TypeFlag = 0
        for index, TypeRow in Types.iterrows():
            if(str(TypeRow['Name'])== currentName):
                TypeFlag = 1
                PokemonString+=str(TypeRow['Type'])+","
            if(str(TypeRow['Name']) != currentName and TypeFlag == 1):
                if PokemonString.split(",")[1] == "":
                    PokemonString+=str("None")+","
                break
        
        for index, ItemRow in Items.iterrows():
            if(str(ItemRow['Name'])== currentName):
                PokemonString+=str(ItemRow['Item'])+","
                break
        #for index, SpreadsRow in Names.iterrows():
         #   if(str(SpreadsRow['Name'])== currentName):
          #      PokemonString+=str(SpreadsRow['Nature'])+","
           #     PokemonString+=str(SpreadsRow['IVs']).replace("/",",")+","
            #    break
        PokemonString+=str(NameRow['Nature'])+","
        PokemonString+=str(NameRow['IVs']).replace("/",",")+","
        for index, AbilitiesRow in Abilities.iterrows():
            if(str(AbilitiesRow['Pokemon'])== currentName):
                PokemonString+=str(AbilitiesRow['Ability'])+","
                break

        Movepool = []
        for index, MovesRow in Moves.iterrows():
            if(str(MovesRow['Pokemon'])== currentName and [str(MovesRow['Move']),(str(MovesRow['Use_Percentage']).replace("%",""))] not in Movepool):
                Movepool.append([str(MovesRow['Move']),(str(MovesRow['Use_Percentage']).replace("%",""))])
                #print([str(MovesRow['Move']),(str(MovesRow['Use_Percentage']).replace("%",""))])
        topFour = []
        #print(Movepool)
        for i in Movepool:
            
            if len(topFour) < 4 and i not in topFour:
                topFour.append(i)
            else:
                for j in topFour:
                    if i[1] > j[1] and i not in topFour:
                        topFour.remove(j)
                        topFour.append(i)
        if len(topFour) == 4:
            for l in topFour:
                PokemonString+=str(l[0])+","
        else:
            for l in topFour:
                PokemonString+=str(l[0])+","
            for i in range(0,4-len(topFour)):
                if len(topFour) != 4:
                    PokemonString+=str("None")+","

                
        
    
        f.write(PokemonString+"\n")#.replace("é","e").replace("�","e"))
```
